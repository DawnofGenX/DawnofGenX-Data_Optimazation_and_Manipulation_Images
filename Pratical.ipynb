```python
ret,thresh1 = cv2.threshold(grey_flower_image,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(grey_flower_image,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(grey_flower_image,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(grey_flower_image,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(grey_flower_image,127,255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [grey_flower_image, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
plt.show()
```

```python
blur = cv2.blur(grey_flower_image,(20,20))
Gaussian_blur = cv2.GaussianBlur(grey_flower_image,(101,101),cv2.BORDER_DEFAULT)

plt.subplot(2,3,1),plt.imshow(grey_flower_image,"gray"),plt.title('Original')
plt.subplot(2,3,2),plt.imshow(blur,"gray"),plt.title('Blurred')
plt.subplot(2,3,3),plt.imshow(Gaussian_blur,"gray"),plt.title('Gaussian Blurred')
plt.show()
```
